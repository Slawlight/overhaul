# Empirical Study on CIFAR-10

Let's play with the CIFAR-10 ([Krizhevsky & Hinton, 2009](https://www.cs.toronto.edu/~kriz/learning-features-2009-TR.pdf)) image classification task.
We will train two benchmark models: ResNet ([He et al., 2016](https://arxiv.org/abs/1512.03385)) and DenseNet ([Huang et al., 2017](https://arxiv.org/abs/1608.06993)) with AdaBound/AMSBound and other baseline methods: SGD, AdaGrad, Adam, and AMSGrad.

We've already provided the code for training and evaluation [here](https://github.com/Luolc/AdaBound/tree/master/demos/cifar10).
We employ the fixed budget of 200 epochs and reduce the learning rates by 10 after 150 epochs.
The online version of this notebook will use the data points of pretrained models, which were produced by AdaBound/AMSBound with default settings and baseline optimizers with their best hyperparameters.
To run the experiments by yourself, please follow the [instructions](https://github.com/Luolc/AdaBound/tree/master/demos/cifar10) and change the plotting parameter in this notebook correspodingly.

```python
import os
%matplotlib notebook
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
def get_folder_path(dataset, top = 1):
    path = 'ckpt/curve/cifar10'
    if dataset == 'cifar100':
        path = 'ckpt/curve/cifar100'
    if dataset == 'imagenet':
        path = 'ckpt/curve/ImageNet/top1'
        if top == 5:
            path = 'ckpt/curve/ImageNet/top5'
    return path
```

```python
def get_curve_data(dataset, model='resnet18', top=1):
    label = ['vanilla2', 'SEF3',  'SE', 'SRM']
    folder_path = get_folder_path(dataset,top=top)
    filenames = [name for name in os.listdir(folder_path) if (model in name)and ('sgd') not in name and ('fixed') in name]
    paths = [os.path.join(folder_path, name) for name in filenames]
   
    keys = [name.split('-')[1]+'-'+ name.split('-')[-2] for name in filenames]
    
    if dataset == 'cifar100':
        filenames = [name for name in os.listdir(folder_path) if (model in name)and ('student') not in name]
        paths = [os.path.join(folder_path, name) for name in filenames]
        keys = [name.split('-')[1]+'-'+ name.split('-')[-1] for name in filenames]
    
    return {key: torch.load(fp) for key, fp in zip(keys, paths)}
```

```python
def plot(model='resnet', optimizers=None, curve_type='train', dataset='cifar10'):
    
    curve_data = get_curve_data(dataset, model=model)
    
    datanum = 10
    if dataset=='cifar100':
        datanum = 100
    
    plt.figure()
    plt.title('{} Accuracy for {} on CIFAR-{}'.format(curve_type.capitalize(), model, datanum))
    plt.xlabel('Epoch')
    plt.ylabel('{} Accuracy %'.format(curve_type.capitalize()))
    
    if dataset == 'cifar10'and curve_type == 'train':
        plt.ylim(80, 101)
    elif dataset == 'cifar10'and curve_type == 'test':
        plt.ylim(80, 96)
    elif dataset == 'cifar100'and curve_type == 'train':
        plt.ylim(40, 100)
    elif dataset == 'cifar100' and curve_type == 'test':
        plt.ylim(40, 80)
    else:
        plt.ylim(30, 100)
    

    for key in curve_data:
        linestyle = '-' 
        accuracies = np.array(curve_data[key]['{}_acc'.format(curve_type)])
        plt.plot(accuracies, label=key, ls=linestyle)
        
    plt.grid(ls='--')
    plt.legend()
    plt.show()
   # plt.savefig('2.png', dpi=200)
```

```python
def plot_imagenet(model='ResNet', curve_type='train', top=1):
    
    curve_data = get_curve_data('imagenet', model=model, top=top)
    
    plt.figure()
    plt.title('{} Error for {} on ImageNet'.format(curve_type.capitalize(), model))
    plt.xlabel('Epoch')
    plt.ylabel('{} Error %'.format(curve_type.capitalize()))
     
    if top == 5:
        plt.ylim(0, 50)
    elif top == 1:
        plt.ylim(30, 75)
    
    for key in curve_data:
        linestyle = '-' 
        accuracies = np.array(curve_data[key]['{}_acc'.format(curve_type)])
        plt.plot(accuracies, label=key, ls=linestyle)
        
    plt.grid(ls='--')
    plt.legend()
    plt.show()
```

The above function is for plotting the learning curves. To use your own data points, set `use_pretrained` as `False`.

## ResNet

First, let's see the results with ResNet-34.

```python
plot(model='resnet34', curve_type='train',dataset='cifar10')
plot(model='resnet34', curve_type='test',dataset='cifar10')
```

```python
plot(model='resnet', curve_type='train',dataset='cifar100')
plot(model='resnet', curve_type='test',dataset='cifar100')
```

```python
plot_imagenet(model='resnet', curve_type='train', top=1)
plot_imagenet(model='resnet', curve_type='val', top=1)
```

We can see that adaptive methods such as AdaGrad, Adam and AMSGrad appear to perform better than the non-adaptive ones early in training. 
But by epoch 150 when the learning rates are decayed, SGD begins to outperform those adaptive methods. 
As for our methods, AdaBound and AMSBound, they converge as fast as adaptive ones and achieve a higher accuracy than SGD on the test set at the end of training. 
In addition, compared with their prototypes, their performances on the unseen data (test set) are enhanced evidently.

You may observe a slight decline of the test accuracy with SGD.
Since the decline also occurs in training, this may not be caused by overfitting but some other reasons.
This behavior indicates the unstable performance of SGD to some extent, which is similar to the observation reported in [He et al. (2016)](https://arxiv.org/abs/1512.03385).
